```python
#Importar os dados
import pandas as pd
caminho = 'C:/Users/guihb/Desktop/aula/eda_aula/res/database/data/SE/DNSE2017.csv'

engdados = pd.read_csv(caminho, dtype=str)

#transformacao dos dados
engdados.PESO = pd.to_numeric(engdados.PESO)
engdados.IDADEMAE = pd.to_numeric(engdados.IDADEMAE).astype(pd.Int64Dtype())
engdados.IDADEPAI = pd.to_numeric(engdados.IDADEPAI).astype(pd.Int64Dtype())
engdados.QTDFILVIVO = pd.to_numeric(engdados.QTDFILVIVO).astype(pd.Int64Dtype())
engdados.QTDFILMORT = pd.to_numeric(engdados.QTDFILMORT).astype(pd.Int64Dtype())
engdados.APGAR1 = pd.to_numeric(engdados.APGAR1).astype(pd.Int64Dtype())
engdados.APGAR5 = pd.to_numeric(engdados.APGAR5).astype(pd.Int64Dtype())
engdados.SEMAGESTAC = pd.to_numeric(engdados.SEMAGESTAC).astype(pd.Int64Dtype())
engdados.CONSPRENAT = pd.to_numeric(engdados.CONSPRENAT).astype(pd.Int64Dtype())
engdados.SERIESCMAE = pd.to_numeric(engdados.SERIESCMAE).astype(pd.Int64Dtype())
engdados.QTDPARTCES = pd.to_numeric(engdados.QTDPARTCES).astype(pd.Int64Dtype())
engdados.QTDPARTNOR = pd.to_numeric(engdados.QTDPARTNOR).astype(pd.Int64Dtype())
engdados.QTDGESTANT = pd.to_numeric(engdados.QTDGESTANT).astype(pd.Int64Dtype())

engdados
```

```python
#descrever a estrutura
engdados.describe()
```

```python
#1. Qual a média da idade da mãe, peso, apgar 1 min, apgar 5 min, quantidade de filhos vivos e quantidade de filhos mortos?

media = engdados[['IDADEMAE','PESO','APGAR1', 'APGAR5', 'QTDFILVIVO', 'QTDFILMORT']].mean()
print(media)
```

```python
#2. Qual a idade tal que metade das idades das mães são menores que ela (mediana)?
mediana = engdados.IDADEMAE.median()
print(mediana)
```

```python
#3. Qual o apgar de 1 min que mais aparece (moda)?
engdados.APGAR1.mode()

```

```python
#4. Qual o apgar de 5 minutos que mais aparece?
engdados.APGAR5.mode()
```

```python
#5. Qual o peso da criança tal que 10% dos pesos são menores (P10)?
engdados.PESO.quantile([0.10])
```

```python
#6. Qual o peso da criança tal que 75% dos pesos são menores (Q3)?
engdados.PESO.quantile([0.75])
```

```python
#7. Qual a quantidade de filhos mortos que mais aparece?
engdados.QTDFILMORT.mode()
```

```python
#8. Qual a média de idade de mães solteiras?
maesolteiras = engdados[engdados.ESTCIVMAE == '1']
maesolteiras.IDADEMAE.mean()
```

```python
#9. Qual a quantidade de filhos vivos das mães com até 3 anos de escolaridade tal que 25% dessa quantidade são menores (Q1)?
escolaridade = engdados[(engdados.ESCMAE == '1') | (engdados.ESCMAE == '2')]
escolaridade.QTDFILVIVO.quantile([0.25])
```

```python
#10. Qual a moda da idade das mães para as crianças que nasceram com anomalia congênita?
anoma = engdados[engdados.IDANOMAL == '1']
anoma.IDADEMAE.mode()
```
