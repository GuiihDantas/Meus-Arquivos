```python
#Importar os dados
import pandas as pd
caminho = 'C:/Users/guihb/Desktop/aula/eda_aula/res/database/data/SE/DNSE2017.csv'

engdados = pd.read_csv(caminho, dtype=str)

#transformacao dos dados
engdados.PESO = pd.to_numeric(engdados.PESO)
engdados.IDADEMAE = pd.to_numeric(engdados.IDADEMAE).astype(pd.Int64Dtype())
engdados.IDADEPAI = pd.to_numeric(engdados.IDADEPAI).astype(pd.Int64Dtype())
engdados.QTDFILVIVO = pd.to_numeric(engdados.QTDFILVIVO).astype(pd.Int64Dtype())
engdados.QTDFILMORT = pd.to_numeric(engdados.QTDFILMORT).astype(pd.Int64Dtype())
engdados.APGAR1 = pd.to_numeric(engdados.APGAR1).astype(pd.Int64Dtype())
engdados.APGAR5 = pd.to_numeric(engdados.APGAR5).astype(pd.Int64Dtype())
engdados.SEMAGESTAC = pd.to_numeric(engdados.SEMAGESTAC).astype(pd.Int64Dtype())
engdados.CONSPRENAT = pd.to_numeric(engdados.CONSPRENAT).astype(pd.Int64Dtype())
engdados.SERIESCMAE = pd.to_numeric(engdados.SERIESCMAE).astype(pd.Int64Dtype())
engdados.QTDPARTCES = pd.to_numeric(engdados.QTDPARTCES).astype(pd.Int64Dtype())
engdados.QTDPARTNOR = pd.to_numeric(engdados.QTDPARTNOR).astype(pd.Int64Dtype())
engdados.QTDGESTANT = pd.to_numeric(engdados.QTDGESTANT).astype(pd.Int64Dtype())

engdados
```

```python
engdados.describe()
```

```python
#criando faixas etárias
faixas = engdados['FAIXAS'] = pd.cut(engdados.IDADEMAE,[0,9,14,19,24,29,34,39,44,49,54,59,64])
#engdados[['IDADEMAE','FAIXAS']]
faixas.head(15)
```

```python
#analizando a quantidade das faixas etárias das mães
faixa_etaria = engdados['IDADEMAE'].groupby(faixas).agg(['count'])
faixa_etaria
```

```python
#1. Fazer o boxplot para idade da mãe e encontrar quais os registro de outliers.
bp = engdados.boxplot(column = ['IDADEMAE'] , figsize = [20,8])
```

```python
1. #Continuação exercício 1
#Limite superior: q3 + 1.5 * (q3 - q1)
#Limite inferior: q3 - 1.5 * (q3 - q1)

q = engdados.IDADEMAE.quantile([0.25 , 0.75]).values
limit_inf = q[0] - 1.5 * (q[1] - q[0])
limit_sup = q[1] + 1.5 * (q[1] - q[0])

outliers = engdados[(engdados.IDADEMAE < limit_inf) | (engdados.IDADEMAE > limit_sup)]
outliers
```

```python
#2. Fazer o boxplot para o peso da criança e encontrar quais os registro de outliers.
bp = engdados.boxplot(column = ['PESO'] , figsize = [20,8])
```

```python
2. #Continuação exercício 2
#Limite superior: q3 + 1.5 * (q3 - q1)
#Limite inferior: q3 - 1.5 * (q3 - q1)

peso = engdados.PESO.quantile([0.25 , 0.75]).values
limit_inf_peso = peso[0] - 1.5 * (peso[1] - peso[0])
limit_sup_peso = peso[1] + 1.5 * (peso[1] - peso[0])

outliers = engdados[(engdados.IDADEMAE < limit_inf_peso) | (engdados.IDADEMAE > limit_sup_peso)]
outliers
```

```python
#3. Fazer o histograma da idade do pai
bins_values = engdados.IDADEPAI.value_counts(sort=False)
hist = engdados.hist(column='IDADEPAI' , bins = sorted(bins_values.keys()) , figsize=(20,8), grid=False, width = 0.8, color='green')

for key, value in bins_values.items():
    hist[0][0].text(key+0.2, value+5, value, ha='center', color ='red')
```

```python
#4. Representar visualmente os diferentes valores de APGAR1 e APGAR5
ap1 = engdados[(engdados.APGAR1 != 8) & (engdados.APGAR1 != 9)].groupby(by='APGAR1').count()
pizza = ap1[['contador']].plot(y='contador',  kind='pie', autopct='%1.1f%%', figsize=(20,8))
r = pizza.set(xlabel='', ylabel='')

ap2 = engdados[(engdados.APGAR1 < 9)].groupby(by='APGAR5').count()
pizza = ap2[['contador']].plot(y='contador',  kind='pie', autopct='%1.1f%%', figsize=(20,8))
r = pizza.set(xlabel='', ylabel='')
```

```python
#5. Mostrar a dispersão entre idade do pai e idade da mãe
idades = engdados[(engdados.IDADEPAI.notna())][['IDADEMAE', 'IDADEPAI']]
disp = idades.plot(kind='scatter',x='IDADEMAE', y='IDADEPAI', figsize=(20,8))
disp.tick_params(colors='white')
disp.xaxis.label.set_color('white')
disp.yaxis.label.set_color('white')
```

```python
#6. Mostrar visualmente a comparação entre idade da mãe e idade do pai para as mães solteiras com menos de 18 anos
idades = engdados[(engdados.IDADEMAE < 18) & (engdados.ESTCIVMAE== '1') & (engdados.IDADEPAI.notna())][['IDADEMAE', 'IDADEPAI']]
barras = idades.plot(kind='bar', width=0.7, figsize=(20,8))
barras.tick_params(colors='white')
barras.xaxis.label.set_color('white')
barras.yaxis.label.set_color('white')
r = barras.set(xlabel='Observação', ylabel='Idade')
```

```python
#7. Comparar a frequência dos anos de escolaridade das mães para que residem na capital (Aracajú, CODMUNRES 2800308) e das mães em outras cidades (outros CODMUNRES)
capital = engdados[engdados.CODMUNRES == '280030']
outras = engdados[engdados.CODMUNRES != '280030']

# gráfico de pizza da capital
grupo = capital.groupby(by='ESCMAE').count()
pizza = grupo[['contador']].plot(y='contador',  kind='pie', autopct='%1.1f%%', figsize=(20,8))
r = pizza.set(xlabel='', ylabel='')

# gráfico de pizza de outras cidades
grupo = outras.groupby(by='ESCMAE').count()
pizza = grupo[['contador']].plot(y='contador',  kind='pie', autopct='%1.1f%%', figsize=(20,8))
r = pizza.set(xlabel='', ylabel='')
```
