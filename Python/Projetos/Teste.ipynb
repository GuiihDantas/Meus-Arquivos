```python
import pandas as pd
```

```python
file = ('temp.csv')
df = pd.read_csv(file, sep=',', encoding='gbk')
df.head(5)
```

```python
df_media = df[['Area','Months','Y1961']].groupby(['Area','Months'], as_index = False).mean()
df_media
```

```python
df_desvio = df[['Area','Y2014']].groupby(['Area'], as_index = False).std(ddof=1)
df_desvio
```

```python
df_desvio["Y2014"] = df_desvio["Y2014"].fillna(0)
df_desvio.head(30)
```
