```python
Utilizando média e medidas de dispersão, faça o seguinte para a base de nascidos vivos de SE em 2017:

1 .Compare as idades das mães que fizeram cesárea e das mães que fizeram parto normal
2. Compare o peso da criança ao nascer das mães que fizeram cesárea e das mães que fizeram parto normal
3. Compare o peso da criança ao nascer para as diferentes categorias de consultas pré-natal
4. Compare quantidade de filhos vivos e mortos em relação ao tipo de gravidez (única, dupla, tripla ou mais)
5. Compare a quantidade total de gravidezes (filhos vivos + filhos mortos) em relação a escolaridade
```

```python
#Importar os dados
import pandas as pd
caminho = 'C:/Users/guihb/Desktop/aula/eda_aula/res/database/data/SE/DNSE2017.csv'

engdados = pd.read_csv(caminho, dtype=str)

#transformacao dos dados
engdados.PESO = pd.to_numeric(engdados.PESO)
engdados.IDADEMAE = pd.to_numeric(engdados.IDADEMAE).astype(pd.Int64Dtype())
engdados.IDADEPAI = pd.to_numeric(engdados.IDADEPAI).astype(pd.Int64Dtype())
engdados.QTDFILVIVO = pd.to_numeric(engdados.QTDFILVIVO).astype(pd.Int64Dtype())
engdados.QTDFILMORT = pd.to_numeric(engdados.QTDFILMORT).astype(pd.Int64Dtype())
engdados.APGAR1 = pd.to_numeric(engdados.APGAR1).astype(pd.Int64Dtype())
engdados.APGAR5 = pd.to_numeric(engdados.APGAR5).astype(pd.Int64Dtype())
engdados.SEMAGESTAC = pd.to_numeric(engdados.SEMAGESTAC).astype(pd.Int64Dtype())
engdados.CONSPRENAT = pd.to_numeric(engdados.CONSPRENAT).astype(pd.Int64Dtype())
engdados.SERIESCMAE = pd.to_numeric(engdados.SERIESCMAE).astype(pd.Int64Dtype())
engdados.QTDPARTCES = pd.to_numeric(engdados.QTDPARTCES).astype(pd.Int64Dtype())
engdados.QTDPARTNOR = pd.to_numeric(engdados.QTDPARTNOR).astype(pd.Int64Dtype())
engdados.QTDGESTANT = pd.to_numeric(engdados.QTDGESTANT).astype(pd.Int64Dtype())

engdados
```

```python
engdados.describe()
```

```python
#1 .Compare as idades das mães que fizeram cesárea e das mães que fizeram parto normal
normal = engdados[engdados.PARTO == '1'][['IDADEMAE']]
cesarea = engdados[engdados.PARTO == '2'][['IDADEMAE']]
media_normal = normal.mean()
media_cesarea = cesarea.mean()
print(media_normal)
print(media_cesarea)
```

```python
#2. Compare o peso da criança ao nascer das mães que fizeram cesárea e das mães que fizeram parto normal
```

```python
#PLUS (DESVIO)
dm_normal = normal.mad()
dm_cesarea = cesarea.mad()
print(dm_normal)
print(dm_cesarea)
```

```python
#PLUS (DESVIO PADRÃO)
dp_normal = normal.std(ddof=0)
dp_cesarea = cesarea.std(ddof=0)
print(dp_normal)
print(dp_cesarea)
```

```python
#PLUS (AMPLITUDE INTERQUARTIL)
q_normal = normal.quantile([0.25,0.75])
q_cesarea = cesarea.quantile([0.25,0.75])
print(q_normal)
print(q_cesarea)
```

```python
#PLUS (COEFICIENTE DE VARIAÇÃO)
cv_normal = dp_normal/media_normal
cv_cesarea = dp_cesarea/media_cesarea
print(cv_normal)
print(cv_cesarea)
```

```python
#2. Compare o peso da criança ao nascer das mães que fizeram cesárea e das mães que fizeram parto normal

normal = engdados[engdados.PARTO == '1'][['PESO']]
cesarea = engdados[engdados.PARTO == '2'][['PESO']]
media_normal = normal.mean()
media_cesarea = cesarea.mean()
print(media_normal)
print(media_cesarea)
```

```python
#3. Compare o peso da criança ao nascer para as diferentes categorias de consultas pré-natal
natal_1a3 = engdados[engdados.CONSULTAS == '1'][['PESO']]
natal_4a6 = engdados[engdados.CONSULTAS == '2'][['PESO']]
natal_7 = engdados[engdados.CONSULTAS == '3'][['PESO']]
media_1a3 = natal_1a3.mean()
media_4a6 = natal_4a6.mean()
media_7 = natal_7.mean()
print(media_1a3)
print(media_4a6)
print(media_7)
```

```python
#4. Compare quantidade de filhos vivos e mortos em relação ao tipo de gravidez (única, dupla, tripla ou mais)
unica_vivo = engdados[engdados.GRAVIDEZ == '1'][['QTDFILVIVO']].mean()
dupla_vivo = engdados[engdados.GRAVIDEZ == '2'][['QTDFILVIVO']].mean()
tripla_vivo = engdados[engdados.GRAVIDEZ == '3'][['QTDFILVIVO']].mean()
unica_morto = engdados[engdados.GRAVIDEZ == '1'][['QTDFILMORT']].mean()
dupla_morto = engdados[engdados.GRAVIDEZ == '2'][['QTDFILMORT']].mean()
tripla_morto = engdados[engdados.GRAVIDEZ == '3'][['QTDFILMORT']].mean()
print(unica_vivo)
print(dupla_vivo)
print(tripla_vivo)
print(unica_morto)
print(dupla_morto)
print(tripla_morto)
```

```python
#Criando uma nova coluna com a soma dos filhos vivos e mortos
engdados['FILHOS'] = engdados.QTDFILVIVO + engdados.QTDFILMORT
engdados[['FILHOS']]
```

```python
#5. Compare a quantidade total de gravidezes (filhos vivos + filhos mortos) em relação a escolaridade

escolaridade_0 = engdados[engdados.ESCMAE == '1'][['FILHOS']].mean()
escolaridade_1a3 = engdados[engdados.ESCMAE == '2'][['FILHOS']].mean() 
escolaridade_4a7 = engdados[engdados.ESCMAE == '3'][['FILHOS']].mean() 
escolaridade_8a11 = engdados[engdados.ESCMAE == '4'][['FILHOS']].mean() 
escolaridade_12 = engdados[engdados.ESCMAE == '5'][['FILHOS']].mean() 
print(escolaridade_0)
print(escolaridade_1a3)
print(escolaridade_4a7)
print(escolaridade_8a11)
print(escolaridade_12)
```
